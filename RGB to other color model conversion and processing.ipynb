```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rgb_to_cmy(image):


    img_normalized = image / 255.0

    cmy_img = 1 - img_normalized

    cmy_img = (cmy_img * 255).astype(np.uint8)
    return cmy_img

def display_images(original, cmy):

    original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)



    plt.imshow(original_rgb)
    plt.title('Original RGB Image')
    plt.axis('off')
    plt.show()

    plt.imshow(cv2.cvtColor(cmy, cv2.COLOR_BGR2RGB))
    plt.title('CMY Image')
    plt.axis('off')
    plt.show()

image_path = 'C:/Users/abina/Downloads/home_image.jpg'
rgb_image = cv2.imread(image_path)
cmy_image = rgb_to_cmy(rgb_image)
display_images(rgb_image, cmy_image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def isolate_color_channels(image):

    red_image = image.copy()
    green_image = image.copy()
    blue_image = image.copy()

    red_image[:, :, 1] = 0
    red_image[:, :, 0] = 0


    green_image[:, :, 0] = 0
    green_image[:, :, 2] = 0

    blue_image[:, :, 2] = 0
    blue_image[:, :, 1] = 0

    return red_image, green_image, blue_image

def display_images(original, red, green, blue):

    original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
    red_rgb = cv2.cvtColor(red, cv2.COLOR_BGR2RGB)
    green_rgb = cv2.cvtColor(green, cv2.COLOR_BGR2RGB)
    blue_rgb = cv2.cvtColor(blue, cv2.COLOR_BGR2RGB)



    plt.subplot(2, 2, 1)
    plt.imshow(original_rgb)
    plt.title('Original RGB Image')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(red_rgb)
    plt.title('Red Channel Image')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(green_rgb)
    plt.title('Green Channel Image')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(blue_rgb)
    plt.title('Blue Channel Image')
    plt.axis('off')

    plt.show()

image_path = 'C:/Users/abina/Downloads/home_image.jpg'
rgb_image = cv2.imread(image_path)

red_image, green_image, blue_image = isolate_color_channels(rgb_image)

display_images(rgb_image, red_image, green_image, blue_image)
```

```python


def rgb_to_yiq(image):


    transformation_matrix = np.array([
        [0.299, 0.587, 0.114],
        [0.596, -0.274, -0.322],
        [0.211, -0.523, 0.312]
    ])


    reshaped_image = image.reshape(-1, 3)

    yiq_image = np.dot(reshaped_image, transformation_matrix.T)

    yiq_image = yiq_image.reshape(image.shape)

    return yiq_image

img = cv2.imread("C:/Users/abina/Downloads/home_image.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


yiq_image = rgb_to_yiq(img_rgb)


yiq_image_display = np.clip((yiq_image - yiq_image.min()) / (yiq_image.max() - yiq_image.min()) * 255, 0, 255).astype(np.uint8)

plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Original RGB Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(yiq_image_display)
plt.title('YIQ Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("C:/Users/abina/Downloads/home_image.jpg")


img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(img_hsv)




img_hsv_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

plt.subplot(2,2,1)
plt.imshow(img_hsv_rgb)
plt.title('Original HSV Image')
plt.axis('off')


plt.subplot(2,2,2)
plt.imshow(H, cmap='gray')
plt.title('Hue Channel')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(S, cmap='gray')
plt.title('Saturation Channel')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(V, cmap='gray')
plt.title('Value Channel')
plt.axis('off')


plt.show()
```

```python

image = cv2.imread('C:/Users/abina/Downloads/home_image.jpg')

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

hue_shift = 36


saturation_scale = 1.2
value_scale = 1.1


h, s, v = cv2.split(hsv_image)


h = (h + hue_shift) % 180
s = cv2.multiply(s, saturation_scale)
v = cv2.multiply(v, value_scale)


adjusted_hsv_image = cv2.merge([h, s, v])


adjusted_image = cv2.cvtColor(adjusted_hsv_image, cv2.COLOR_HSV2BGR)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Adjusted Image')
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))

plt.show()


```
